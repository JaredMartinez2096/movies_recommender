```python
#import libraries
import pandas as pd
import numpy as np
import string
```

```python
#Charge de dataset
df = pd.read_csv("Netflix Dataset Latest 2021.csv")
df.head(5)
```

```python
#Show the columns to select the right to its use
df.columns
```

```python
#Drop the unnecessary columns 
Recomen = df.drop(columns=['Genre', 'Languages', 'Series or Movie',
       'Hidden Gem Score', 'Country Availability', 'Runtime', 'Director',
       'Writer', 'Actors', 'View Rating', 'IMDb Score',
       'Rotten Tomatoes Score', 'Metacritic Score', 'Awards Received',
       'Awards Nominated For', 'Boxoffice', 'Release Date',
       'Netflix Release Date', 'Production House', 'Netflix Link', 'IMDb Link',
       'Summary', 'IMDb Votes', 'Image', 'Poster', 'TMDb Trailer',
       'Trailer Site','Unnamed: 29', 'Unnamed: 30'])
```

```python
#Sum the null values 
Recomen.isnull().sum()
```

```python
#Drop the null values because if we don´t have "tags", we would'nt know what kind of movie is and it is a unnecessary movie or serie.
Recomen = Recomen.dropna(axis=0)
Recomen = Recomen.reset_index(drop=True)
```

```python
#Elimiten the unnecessary puntuaction to can use the label in our matrix
Recomen["Clean_Tags"] = Recomen["Tags"].str.replace("[{}]".format(string.punctuation), " ")
Recomen
```

```python
#Import sklearn to use Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words="english")
Recomen["Clean_Tags"] = Recomen["Clean_Tags"].fillna("")
tfidf_matrix = tfidf.fit_transform(Recomen["Clean_Tags"])
print(len(tfidf.vocabulary_))
```

```python
#We create the matrix to comparate the labels
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

indices = pd.Series(Recomen.index, index=Recomen["Title"]).drop_duplicates()
```

```python
#Create a method to get recommendations
def get_recommendations(title, cosine_sim=cosine_sim):
    idx=indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key =lambda x: x[1], reverse = True)
    sim_scores = sim_scores[1:5]
    movie_indices = [i[0] for i in sim_scores]
    return Recomen["Title"].iloc[movie_indices]
```

```python
#We can choose a random id and get a movie and the model recommend us 4 similar movies, it depens of the label
titulo = str(Recomen.iloc[200]["Title"])
print("Tu seleccionaste: " +titulo+ " y tus recomendaciones son: ")
print(get_recommendations(titulo))
```

```python
#We can write de movie and the model recommens 4 movies too.
seleccion = str(input("Escribe tu película: "))
print("Tu selección fue " +seleccion+ " y tus recomendaciones son: ")
print(get_recommendations(seleccion))
```
